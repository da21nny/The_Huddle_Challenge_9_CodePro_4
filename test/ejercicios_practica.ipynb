## Ejercicio 8 - Ordenamiento
Ordena el DataFrame por `votes_received` de **mayor a menor** y muestra los **10 primeros** registros (solo columnas `user_id`, `post_category`, `votes_received`).

## Ejercicio 9 - Crear nueva columna
Crea una columna llamada `es_popular` que sea `True` si `votes_received >= 10`, y `False` en caso contrario. Luego cuenta cuántos posts son populares y cuántos no.

```python
# Tu código aquí
df['es_popular'] = df['votes_received'] >= 10

print(df['es_popular'].value_counts())
```
