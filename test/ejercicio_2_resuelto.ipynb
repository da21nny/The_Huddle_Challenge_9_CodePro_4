## 2 - Últimas filas y muestra

```python
print(df.tail(3))
print('\nMuestra aleatoria:')
print(df.sample(5, random_state=42))
```

---
## 3 - Selección de columnas

```python
df_seleccion = df[['user_id', 'country', 'votes_received']]
df_seleccion.head()
```

---
## 4 - Conteo por país

```python
df['country'].value_counts().sort_values(ascending=False)
```

---
## 5 - Filtro con múltiples condiciones

```python
filtrado = df[(df['country'] == 'US') & (df['votes_received'] > 10)]
print(f'Registros de US con más de 10 votos: {len(filtrado)}')
filtrado.head()
```

---
## 6 - Máximo y mínimo

```python
print(f'Edad máxima: {df["user_age"].max()}')
print(f'Edad mínima: {df["user_age"].min()}')

print('\nUsuario con más votos:')
print(df.loc[df['votes_received'].idxmax()])
```

---
## 7 - Agrupar y contar

```python
df.groupby('post_category').size().sort_values(ascending=False).head(5)
```

---
## 8 - Mediana y suma

```python
print(f'Mediana de days_since_signup: {df["days_since_signup"].median()}')
print(f'Suma total de votes_received: {df["votes_received"].sum()}')
```

---
## 9 - Reemplazar nulos

```python
mediana_edad = df['user_age'].median()
print(f'Mediana de user_age: {mediana_edad}')

df['user_age'] = df['user_age'].fillna(mediana_edad)

print(f'Nulos en user_age después: {df["user_age"].isnull().sum()}')
```

---
## 10 - Usuarios únicos

```python
usuarios_unicos = df['user_id'].nunique()
total_registros = len(df)
promedio = total_registros / usuarios_unicos

print(f'Usuarios únicos: {usuarios_unicos}')
print(f'Registros totales: {total_registros}')
print(f'Promedio de registros por usuario: {promedio:.2f}')
```
