## 6 - Máximo y mínimo

## 8 - Mediana y suma
Calcula la **mediana** de `days_since_signup` y la **suma total** de `votes_received`.

```python
# Tu código aquí
mediana = df['days_since_signup'].median()
suma = df['votes_received'].sum()

print(mediana)
print(suma)
```

---
## 9 - Reemplazar nulos
Reemplaza los nulos de `user_age` con la **mediana** de esa columna. Verifica que ya no haya nulos.

```python
# Tu código aquí
mediana = df['user_age'].median()
print(mediana)
df['user_age'] = df['user_age'].fillna(mediana)
print(f"no nulos despues {df['user_age'].isnull().sum()}")
```

## 10 - Usuarios únicos
¿Cuántos `user_id` **únicos** hay? ¿Cuántos registros totales? Calcula el promedio de registros por usuario.

```python
# Tu código aquí
unico = df['user_id'].nunique()
total = len(df)
promedio = total/unico

print(f'{promedio:.2f}')
```
